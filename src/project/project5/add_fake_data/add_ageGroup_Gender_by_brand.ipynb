```python
import random

brand_info = {
    'Beauty of Joseon': {
        'age_groups': ['18-24', '25-34', '35-44', '45-54', '55+'],
        'age_group_probs': [19.79, 34.57, 16.82, 13.28, 15.54],
        'gender_probs': {'Female': 75.74, 'Male': 24.26},
    },
    'Dr.Jart+': {
        'age_groups': ['18-24', '25-34', '35-44', '45-54', '55+'],
        'age_group_probs': [13.2, 36.77, 15.44, 14.54, 20.05],
        'gender_probs': {'Female': 80, 'Male': 20},
    },

    "I'm from": {
        'age_groups': ['10', '20', '30', '40+'],
        'age_group_probs': [2.50, 42.94, 44.46, 10.10],
        'gender_probs': {'Female': 93.06, 'Male': 6.94},
    },
    'COSRX': {
        'age_groups': ['18-24', '25-34', '35-44', '45-54', '55+'],
        'age_group_probs': [18.39, 33.02, 17.22, 13.19, 18.18],
        'gender_probs': {'Female': 65.59, 'Male': 34.41},
    },
    'PURITO': {
        'age_groups': ['18-24', '25-34', '35-44', '45-54', '55+'],
        'age_group_probs': [17.87, 38.55, 14.12, 11.66, 17.80],
        'gender_probs': {'Female': 83.88, 'Male': 16.12},
    }
}


def assign_age_and_gender(brand, brand_info):
    # 브랜드의 나이대와 성별 확률을 가져옵니다.
    age_groups = brand_info[brand]['age_groups']
    age_group_probs = brand_info[brand]['age_group_probs']
    gender_probs = brand_info[brand]['gender_probs']
    
    # 나이대 랜덤 선택
    age_group = random.choices(age_groups, weights=age_group_probs)[0]
    
    # 성별 랜덤 선택
    gender = random.choices(list(gender_probs.keys()), weights=gender_probs.values())[0]
    
    return age_group, gender
```

```python
# amazon_df의 브랜드 열을 기준으로 나이대와 성별을 추가합니다.
amazon_df['age_group'] = amazon_df['brand'].apply(lambda x: assign_age_and_gender(x, brand_info)[0])
amazon_df['gender'] = amazon_df['brand'].apply(lambda x: assign_age_and_gender(x, brand_info)[1])

# 결과 확인
amazon_df.head()
```
