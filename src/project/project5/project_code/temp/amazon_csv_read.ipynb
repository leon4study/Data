```python
# ASIN이 'B00PBX3L7K'인 행을 필터링
duplicate_rows = itmes_df[itmes_df.ASIN == 'B00PBX3L7K']

# 두 행 간의 차이 확인
difference = duplicate_rows.iloc[0].compare(duplicate_rows.iloc[1])
print(difference)
```

```python
# ASIN 값을 기준으로 먼저 들어간 값 삭제
itmes_df = itmes_df.drop_duplicates(subset=['ASIN'], keep='last')
reviews_df = reviews_df.drop_duplicates(subset=['review_num'], keep='last')
```
