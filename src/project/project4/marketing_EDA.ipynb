### 1. 북극성 지표(North Star Metric)와 가드레일 지표(Guardrail Metric)
>가드레일 지표: 북극성 지표를 뒷받침하며 사업 운영이나 성과에 부정적인 영향을 주지 않도록 보조하는 지표입니다.

북극성 지표
매출(Revenue)
    - 이유: '매출'은 마케팅 캠페인의 최종 성과이자 비즈니스의 핵심 목표입니다. 캠페인을 통해 고객이 제품을 구매하도록 유도하는 것이 주 목적이므로 Revenue를 최우선 지표로 삼아야 합니다.

가드레일 지표
북극성 지표를 달성하는 과정에서 주의해야 할 보조 지표들입니다:
1) 전환률(Conversion Rate)
   - 이유: 캠페인이 고객 행동(전환)을 얼마나 효과적으로 유도했는지를 보여줍니다. 높은 전환률 = 타겟팅 및 메시지 전략이 적절하다는 의미입니다.
   - 예시: 전환률이 너무 낮다면 캠페인 목표와 고객 타겟이 맞지 않을 가능성이 있습니다.

2) ROI (Return on Investment)
   - 이유: 투입된 비용 대비 수익을 측정합니다. ROI가 낮다면 비용 대비 성과가 비효율적이라는 의미입니다.

3) 캠페인 비용(Cost)
   - 이유: 예산 초과 없이 효율적으로 캠페인을 운영하는지 확인합니다.

4) 클릭수(Clicks)와 참여율(Engagement Rate)
   - 이유: 고객이 캠페인 콘텐츠에 얼마나 반응하고 있는지 측정합니다. 참여율이 낮다면 콘텐츠 개선이 필요할 수 있습니다.

5) 노출 횟수(Impressions)
   - 이유: 캠페인이 얼마나 많은 사람에게 도달했는지 측정합니다. Awareness 캠페인의 주요 성과 지표입니다.

</br>


### 2. 분석 주제 및 접근 방법
#### 주제 1: 현 상태 확인 (마케팅 성과 리뷰)
1) 목표: 캠페인 전체 성과를 확인하고 주요 Metric 추세를 분석합니다.
2) 주요 분석 내용
   - Revenue(매출)와 ROI(투자 대비 수익)의 트렌드 분석
   - 캠페인별 전환률 및 참여율 비교
   - 비용(Cost) 대비 성과(Revenue, Conversion) 분석
   - 국가별, 채널별 성과 비교
3) 분석 예시:
   - 가장 높은 ROI를 기록한 캠페인과 가장 낮은 ROI 캠페인 비교
   - 채널별(예: Facebook, Social Media 등) 클릭수와 전환률 분석


#### 주제 2: 사업 수립 방향 결정
1) 목표: 캠페인별 효과를 평가하고 미래 마케팅 전략에 활용합니다.
2) 주요 분석 내용:
   - **캠페인 목표(Campaign Objective)**에 따른 성과 분석(예: Conversion, Awareness, Engagement)
   - 전환 퍼널 단계별 성과 비교
   - 비용 대비 성과가 가장 좋은 타겟 고객군 확인

3) 분석 예시:
   - Conversion 목표 캠페인이 실제로 매출에 가장 큰 기여를 했는가?
   - 특정 국가나 OS(iOS/Android)에서 더 높은 성과를 보이는가?

#### 주제 3: 고객 세그먼트별 인사이트 도출
1) 목표: 고객 타겟별로 어떤 전략이 효과적이었는지 확인합니다.

2) 주요 분석 내용:
   - Target Audience(연령대)에 따른 전환률 및 참여율 비교
   - 국가별, OS별 고객 행동 차이 분석
   - 캠페인 목표별 주요 타겟 고객군 확인

3) 분석 예시:
   - 45-54 연령대가 Conversion Rate에서 높은 성과를 보이는가?
   - 특정 OS(iOS)가 전환율이 더 높은가?

#### 주제 4: 이슈 및 특이사항 확인
1) 목표: 성과가 저조하거나 예상과 다른 캠페인을 발견하고 원인을 분석합니다.

2) 주요 분석 내용: 
   - ROI가 극단적으로 낮은 캠페인 탐색
   - 특정 기간에 노출 대비 클릭수(CTR)가 현저히 낮은 경우
   - 국가나 채널별로 성과가 편차를 보이는 캠페인 확인

3) 분석 예시:
   - 예산은 높았지만 전환이 적었던 캠페인의 원인 분석
   - 특정 시점에서의 이슈: 예를 들어 클릭수 급감, 비용 급증 등

</br>

### 3. 데이터 전처리 주요 포인트
시각화 및 분석 전 다음과 같은 데이터 전처리가 필요합니다:

1) 결측치 처리:
   - Cost, Clicks, Revenue 등의 값이 없는 경우 확인 및 처리

2) 중복 데이터 제거:
   - 동일한 Campaign ID로 중복된 데이터 확인

3) 데이터 타입 변환:
   - 날짜 컬럼 Activity Date를 datetime 형식으로 변환

4) 파생 변수 생성:
   - 클릭당 비용(Cost per Click, CPC) = Cost / Clicks
   - 고객당 매출(Revenue per Conversion, RPC) = Revenue / Conversions

</br>

```python
import pandas as pd

# marketing_df 데이터프레임 복사
df1 = marketing_df.copy()

# 월별로 그룹화하여 합계 계산
monthly_data = df1.groupby('month').agg({
    'Impressions': 'sum',
    'Clicks': 'sum',
    'Conversions': 'sum'
}).reset_index()

# 클릭률과 전환율 계산 (각 월별)
monthly_data['click_through_rate'] = monthly_data['Clicks'] / monthly_data['Impressions']
monthly_data['conversion_rate'] = monthly_data['Conversions'] / monthly_data['Clicks']

# Sankey Chart용 데이터 생성
sankey_data = []

for _, row in monthly_data.iterrows():
    # Impressions -> Clicks
    sankey_data.append({
        'source': 'Impressions',
        'target': 'Clicks',
        'month': row['month'],
        'value': row['Clicks']  # 클릭 수
    })
    # Clicks -> Conversions
    sankey_data.append({
        'source': 'Clicks',
        'target': 'Conversions',
        'month': row['month'],
        'value': row['Conversions']  # 전환 수
    })

# Sankey 데이터프레임 생성
sankey_df = pd.DataFrame(sankey_data)

# Sankey 데이터 저장
sankey_df.to_csv('sankey_chart_data.csv', index=False)

print("Sankey 데이터 준비 완료! Tableau에서 'sankey_chart_data.csv'를 사용하세요.")
```

```python
import pandas as pd

# 예제 데이터


# Sankey 데이터를 구성하기 위한 리스트
sankey_data = []

for _, row in monthly_data.iterrows():
    impressions = row["Impressions"]
    clicks = row["Clicks"]
    conversions = row["Conversions"]
    month = row['month']

    # Level 1: Impressions
    sankey_data.append({
        "month": month,
        "source": "Start",
        "target": "Impressions",
        "values": impressions,
        "out": 0,
        "gone": 0
    })

    # Level 2: Clicks
    sankey_data.append({
        "month": month,
        "source": "Impressions",
        "target": "Clicks",
        "values": clicks,
        "out": impressions - clicks,
        "gone": 0
    })

    # Level 3: Conversions
    sankey_data.append({
        "month": month,
        "source": "Clicks",
        "target": "Conversions",
        "values": conversions,
        "out": clicks - conversions,
        "gone": impressions - clicks
    })


# Sankey용 DataFrame 생성
sankey_df = pd.DataFrame(sankey_data)

# 결과 확인
sankey_df
```
