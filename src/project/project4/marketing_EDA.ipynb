```python
import pandas as pd

# marketing_df ë°ì´í„°í”„ë ˆì„ ë³µì‚¬
df1 = marketing_df.copy()

# ì›”ë³„ë¡œ ê·¸ë£¹í™”í•˜ì—¬ í•©ê³„ ê³„ì‚°
monthly_data = df1.groupby('month').agg({
    'Impressions': 'sum',
    'Clicks': 'sum',
    'Conversions': 'sum'
}).reset_index()

# í´ë¦­ë¥ ê³¼ ì „í™˜ìœ¨ ê³„ì‚° (ê° ì›”ë³„)
monthly_data['click_through_rate'] = monthly_data['Clicks'] / monthly_data['Impressions']
monthly_data['conversion_rate'] = monthly_data['Conversions'] / monthly_data['Clicks']

# Sankey Chartìš© ë°ì´í„° ìƒì„±
sankey_data = []

for _, row in monthly_data.iterrows():
    # Impressions -> Clicks
    sankey_data.append({
        'source': 'Impressions',
        'target': 'Clicks',
        'month': row['month'],
        'value': row['Clicks']  # í´ë¦­ ìˆ˜
    })
    # Clicks -> Conversions
    sankey_data.append({
        'source': 'Clicks',
        'target': 'Conversions',
        'month': row['month'],
        'value': row['Conversions']  # ì „í™˜ ìˆ˜
    })

# Sankey ë°ì´í„°í”„ë ˆì„ ìƒì„±
sankey_df = pd.DataFrame(sankey_data)

# Sankey ë°ì´í„° ì €ì¥
sankey_df.to_csv('sankey_chart_data.csv', index=False)

print("Sankey ë°ì´í„° ì¤€ë¹„ ì™„ë£Œ! Tableauì—ì„œ 'sankey_chart_data.csv'ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.")
```

```python
import pandas as pd

# ì˜ˆì œ ë°ì´í„°


# Sankey ë°ì´í„°ë¥¼ êµ¬ì„±í•˜ê¸° ìœ„í•œ ë¦¬ìŠ¤íŠ¸
sankey_data = []

for _, row in monthly_data.iterrows():
    impressions = row["Impressions"]
    clicks = row["Clicks"]
    conversions = row["Conversions"]
    month = row['month']

    # Level 1: Impressions
    sankey_data.append({
        "month": month,
        "source": "Start",
        "target": "Impressions",
        "values": impressions,
        "out": 0,
        "gone": 0
    })

    # Level 2: Clicks
    sankey_data.append({
        "month": month,
        "source": "Impressions",
        "target": "Clicks",
        "values": clicks,
        "out": impressions - clicks,
        "gone": 0
    })

    # Level 3: Conversions
    sankey_data.append({
        "month": month,
        "source": "Clicks",
        "target": "Conversions",
        "values": conversions,
        "out": clicks - conversions,
        "gone": impressions - clicks
    })


# Sankeyìš© DataFrame ìƒì„±
sankey_df = pd.DataFrame(sankey_data)

# ê²°ê³¼ í™•ì¸
sankey_df
```
