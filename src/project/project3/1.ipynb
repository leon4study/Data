```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 파일 경로 설정
data_path = "K:\\SPARTA\\학습자료\\Python\\marketing_project"

# 데이터 로드
customers = pd.read_csv(os.path.join(data_path, "customers.csv"))
order_items = pd.read_csv(os.path.join(data_path, "order_items.csv"))
orders = pd.read_csv(os.path.join(data_path, "orders.csv"))
payments = pd.read_csv(os.path.join(data_path, "payments.csv"))
products = pd.read_csv(os.path.join(data_path, "products.csv"))

def plot_distribution(df, column_name):

    plt.figure(figsize=(10, 5))
    
    #숫자형 데이터 처리
    if pd.api.types.is_numeric_dtype(df[column_name]):
        sns.histplot(df[column_name].dropna(), kde=True, bins=30, color='blue')
        plt.title(f"Distribution of {column_name} (Numeric)")
        plt.xlabel(column_name)
        plt.ylabel("Frequency")
    #범주형 데이터 처리    
    elif pd.api.types.is_categorical_dtype(df[column_name]) or df[column_name].dtype == 'object':
        value_counts = df[column_name].value_counts().head(20)  # 상위 20개 값만 표시
        sns.barplot(x=value_counts.values, y=value_counts.index, palette='viridis')
        plt.title(f"Top 20 Categories in {column_name}")
        plt.xlabel("Count")
        plt.ylabel(column_name)
    #기타 데이터 처리
    else:
        plt.text(0.5, 0.5, f"Cannot plot column: {column_name}", 
                 horizontalalignment='center', verticalalignment='center', fontsize=12)
        plt.title(f"Unsupported data type for {column_name}")
    
    plt.tight_layout()
    plt.show()

# 데이터 모든 컬럼 시각화
def plot_all_columns(df, df_name):
    print(f"\nVisualizing columns in {df_name} DataFrame:\n")
    for column in df.columns:
        print(f"Visualizing {column}...")
        plot_distribution(df, column)

# 각 데이터프레임의 컬럼 분포 시각화
plot_all_columns(customers, "customers")
plot_all_columns(order_items, "order_items")
plot_all_columns(orders, "orders")
plot_all_columns(payments, "payments")
plot_all_columns(products, "products")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import LabelEncoder

# 데이터 경로 설정
data_path = "K:/SPARTA/학습자료/Python/marketing_project"

# 데이터 로드
customers = pd.read_csv(os.path.join(data_path, "customers.csv"))
order_items = pd.read_csv(os.path.join(data_path, "order_items.csv"))
orders = pd.read_csv(os.path.join(data_path, "orders.csv"))
payments = pd.read_csv(os.path.join(data_path, "payments.csv"))
products = pd.read_csv(os.path.join(data_path, "products.csv"))

# 데이터 병합
df = pd.merge(orders, customers, on="customer_id", how="left")
df = pd.merge(df, order_items, on="order_id", how="left")
df = pd.merge(df, payments, on="order_id", how="left")
df = pd.merge(df, products, on="product_id", how="left")

# 비숫자형 컬럼 처리
for col in df.select_dtypes(include=['object']).columns:
    df[col] = LabelEncoder().fit_transform(df[col].astype(str))

# 상관계수 계산
correlation_matrix = df.corr()

# 히트맵 시각화
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=False, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap for All Columns (Including Encoded Non-Numeric Columns)", fontsize=16)
plt.show()

# 유의미한 상관계수 필터링 (±0.6 이상)
significant_corr = correlation_matrix[(correlation_matrix >= 0.6) | (correlation_matrix <= -0.6)]

# 유의미한 상관계수 히트맵
plt.figure(figsize=(14, 12))
sns.heatmap(significant_corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Significant Correlation Heatmap (>= 0.6 or <= -0.6)", fontsize=16)
plt.show()
```
