#### <span style="color:blue"> **📘 2-9. 테이블 그루핑** </span>

예제를 통해 SQL 구문과 비교해봅시다!

**[예제1]: df2 테이블을 활용한 gender 기준 customer id count하기**

<br>
<span style="color:blue">
1. SQL 활용<br>
   select Gender, count(Customer ID)<br>
   from df2<br>
   group by Gender<br> </span><br>
<span style="color:red">
2. Python 활용<br>
df2.groupby('Gender')['Customer ID'].count() </span>

```python
#직접 실행해봅시다!
df2.groupby('Gender')['Customer ID'].count()
```

**[예제2]: df2 테이블을 활용한 gender,Location 기준 customer id count하기**

<br>
<span style="color:blue">
1. SQL 활용<br>

----------------------------


**[예제3]: df2 테이블을 활용한 Location 기준 Age distinct count하기**

<br>
<span style="color:blue">
1. SQL 활용<br>
   select Location, count(distinct Age)<br>
   from df2<br>
   group by Location<br> </span><br>
<span style="color:red">
2. Python 활용<br>
df2.groupby('Location')['Age'].nunique() </span>

```python
# count는 중복제거를 하지 않고, unique 는 중복을 제거합니다. :) 
#df2.groupby('Location')['Age'].count()와 비교해 봅시다.! 
df2.groupby('Location')['Age'].nunique()
```

----------------------------

**[예제4]: df2 테이블을 활용한 Location 기준 Age distinct count 및 정렬**

<br>
<span style="color:blue">
1. SQL 활용<br>
   select Location, count(distinct Age) as cnt <br>
   from df2<br>
   group by Location<br>
   order by cnt desc <br> </span><br>
<span style="color:red">
2. Python 활용<br>
df2.groupby('Location')['Age'].count().sort_values(ascending=False) </span>

```python
#sort_valuse 사용시, ascending=True 는 오름차순, False 는 내림차순
df2.groupby('Location')['Age'].count().sort_values(ascending=False)
```
